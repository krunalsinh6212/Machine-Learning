```python
import pandas as pd
```

```python
df = pd.read_csv("E:\\JUPYTER\\Machine Learning\\Na√Øve Bayesian Classifier\\Titanic-Dataset.csv")
```

```python
df.head()
```

```python
df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis = 'columns', inplace = True)
```

```python
df.head()
```

```python
inputs = df.drop('Survived', axis='columns')
target = df['Survived']

# Fix the typo: use 'inputs' instead of 'input'
inputs['Sex'] = inputs['Sex'].map({'male': 1, 'female': 2})
```

```python
dummies = pd.get_dummies(inputs['Sex'])
dummies.head(3)
```

```python
inputs = pd.concat([inputs, dummies], axis = 'columns')
```

```python
inputs.head()
```

```python
inputs.drop(['Sex', 1], axis = 'columns', inplace = True)
```

```python
print(inputs.columns)  # Check available columns
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size = 0.3)
```

```python
x_train.columns = x_train.columns.astype(str)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_train, y_train)
```

```python
x_test.columns = x_test.columns.astype(str)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(), x_train, y_train, cv = 5)
```

```python
from sklearn.naive_bayes import BernoulliNB
model1 = BernoulliNB()
```

```python
model1.fit(x_train, y_train)
```

```python
model1.score(x_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model2 = LogisticRegression()
```

```python
model2.fit(x_train, y_train)
```

```python
model2.score(x_test, y_test)
```
