```python
# import required modules
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
# features names
iris.feature_names
```

```python
# target names
iris.target_names
```

```python
# printing the data
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
# printing the data with traget
df['target']=iris.target
df.head()
```

```python
# printing the data with target == 1
df[df.target==1].head()
```

```python
# printing the data with target == 2
df[df.target==2].head()
```

```python
# printing the data with target == 0 and flower name
df['flower_name']=df.target.apply(lambda x:iris.target_names[x])
df.head()
```

```python
df[45:55]
```

```python
# storing the data according to the classification
df0=df[:50]
df1=df[50:100]
df2=df[100:]
```

```python
# import required modules
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# creating the graph with sepal data
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color="green",marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color="blue",marker='.')
```

```python
# creating the graph with petal data
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color="green",marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color="blue",marker='.')
```

```python
# importing the required module
from sklearn.model_selection import train_test_split
```

```python
# splitting the data
x=df.drop(['target','flower_name'],axis='columns')
y=df.target
```

```python
# training the model
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
# total training data
len(X_train)
```

```python
# importing the required module
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=10)
```

```python
# fit the training data
knn.fit(X_train,y_train)
```

```python
# calculating the test score
knn.score(X_test,y_test)
```

```python
# predict the target value
knn.predict([[4.8,3.0,1.5,0.3]])
```

```python
# creating the confusion mattrix
from sklearn.metrics import confusion_matrix
y_pred=knn.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
# creating the diagram of the confusion mattrix
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
# creating report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
