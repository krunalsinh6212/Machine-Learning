```python
df['New York'] = np.where(df['State']=='New York', 1, 0) 
df['California'] = np.where(df['State']=='California', 1, 0) 
df['Florida'] = np.where(df['State']=='Florida', 1, 0) 
df.drop(columns=['State'], axis=1, inplace=True) 
```

```python
df.head()
```

```python
y='Profit' 
x = df.columns.tolist() 
x.remove(y) 
```

```python
x=df[x].values
```

```python
y=df[y].values 
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression  # Corrected spelling

# Assuming X and y are already defined
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)

# Initialize Linear Regression model
lr = LinearRegression()  # Corrected spelling

# Transforming Data using MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)  # Fit and transform training data
X_test = scaler.transform(X_test)  # Only transform test data to prevent data leakage

X_train  # This will display the transformed training data
```

```python
lr = LinearRegression() 
lr.fit(X_train,y_train) 
```

```python
lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Predicting values
y_pred = lr.predict(X_test)

# Evaluating the model
score = r2_score(y_test, y_pred)  # Corrected `r2_score`
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

# Printing results
print(f'R2 Score: {score}')
print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
```
